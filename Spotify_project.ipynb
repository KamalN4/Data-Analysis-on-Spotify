# Loading important tools

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Importing Data 

```python
tracks = pd .read_csv('tracks.csv')
```

```python
tracks.head(10)
```

This dataset contains 586672 rows and 20 columns

# Preliminary Analysis and Cleaning

## Checking out for null values

```python
pd.isnull(tracks).sum()
```

## Dropping out null values 

```python
tracks = tracks.dropna(axis =0)
pd.isnull(tracks).sum()
```

## Checking out for duplicated values

```python
tracks.duplicated().sum()
```

```python
tracks.info()
```

## Explanation of column names that are used for analysis
**id** - It is the id of that song (also the primary key)<br>
**name** - The name of that song <br>
**popularity** - On a scale of 0 - 100 songs have been classified as per their popularity<br>
**duration_ms** - This column refers to duration of songs in milliseconds which will be further converted in seconds<br>
**artists** - Name of the artist that have worked on that song <br>
**id_artists** - Identifier for artists <br>
**release_date** - Date corresponding to which the song was released<br> 
**danceability** - On a scale of 0-1 it measures the value returned corresponding to the ease with which a person could dance to a song over the course of the whole song.<br>
**energy** - The music is probably loud, rhythmically active, and is the kind of music that gets the blood pumping.<br>
**key** - Key in music is a system of functionally related chords deriving from the major and minor scales, with a central note, called the tonic (or keynote).<br>
**Loudness** -  It refers to how loud or soft a sound seems to a distant listener<br>
**mode** - Music is classified into two modes major and minor which are further sub-divided into 7 .<br>
**Speechiness** -  Speechiness detects the presence of spoken words in a track. e.g. talk show, audio book, poetry<br>
**valence** - Describes the musical positiveness conveyed by a track<br>
**tempo**- Tempo is the Italian word for the speed of a song or piece of music.<br> 
**time_signature** - Time signatures, or meter signatures, indicate how many beats are in each measure of a piece of music, as well as which note value is counted as a beat.<br>

##  Fetching 10 least popular songs

```python
sorted_tracks = tracks.sort_values('popularity' , ascending = True)
```

```python
sorted_tracks.head(10)
```

## Obtaining 10 most popular songs which have popularity value >80

```python
most_popular = tracks[tracks['popularity'] > 80].sort_values('popularity' , ascending = False)
most_popular.head(10)
```

## Grouping songs date-wise

```python
tracks['release_date'] = pd.to_datetime(tracks['release_date'] , format='%Y-%m-%d ')
```

```python
tracks['release_date'].dtype
```

```python
tracks['year'] = tracks['release_date'].dt.year
```

```python
tracks.set_index('release_date' , inplace = True)
```

```python
tracks.head()
```

## Converting duration from milli-seconds to seconds

```python
tracks['duration'] = tracks['duration_ms'].apply(lambda x:round(x/1000) )
tracks.drop('duration_ms', axis = 1 , inplace = True)
```

```python
tracks.head()
```

# Visualization and Analysis

## Taking a representative sample out of dataset for analysis

```python
sample_track = tracks.sample(int(0.004*len(tracks)))
```

```python
print(len(sample_track))
```

## Relation between Loudness and energy on representative data 

```python
plt.figure(figsize = (10,8))
sns.regplot(data = sample_track , y = 'loudness' , x = 'energy' , color = 'b').set(title = 'Loudness versus energy')
```

## Relation Between popularity and accoustiveness

```python
plt.figure(figsize = (10,8))
sns.regplot(data = sample_track , y = 'popularity' , x = 'acousticness' , color = 'c').set(title = 'Popularity versus accousticness')
```

## Count of songs year wise

```python
plt.figure(figsize=(12,10))
sns.set(style = 'darkgrid')
sns.displot(data = tracks , x = 'year' ,kind = 'hist' , bins = 100)
```

# Importing SpotifyFeatures.csv file for analysis

```python
genre = pd.read_csv('SpotifyFeatures.csv')
```

```python
genre.head()
```

## Duration of Songs Genre-wise

```python
plt.figure(figsize = (12,10))
plt.title('Duration of songs in Different genre')
sns.barplot(y = 'genre' , x = 'duration_ms' , data = genre)
plt.xlabel('duration in milli-seconds')
plt.ylabel('Genres')
```

## Top 3 Genres by popularity

```python
plt.figure(figsize=(12,5))
famous = genre.sort_values("popularity" , ascending = False).head(5)
sns.barplot(y = 'genre' , x = 'popularity' , data= famous).set(title="Top 3 genres by popularity")
```

## Understanding variation of popularity with mode

```python
plt.figure(figsize = (12,8))
sns.boxplot(x = 'popularity' , y = 'mode' , data = genre , orient = 'h' )
```
